```python
from sklearn import tree
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
# As features são salário, clt ou não e tempo no cargo
# Cada amostra de X traz (nessa ordem: salário, clt e tempo na empresa )
#clt (sim=1, não=0)
# y (pagou=0, não pagou=1)
#tempo = anos

# Observe que não houve standardization...


X =np.array(
    [[1000, 1,4],
     [2000, 1,5],
     [3000, 0,2],
     [4000, 0,2],
     [5000, 1,1],
     [6000, 0,2],
     [7000, 1,1]]
)   

     
y=np.array([0,1,0,0,1,1,1]) #valor 1 = não pagamento (default)

#treinamento da árvore
clf = tree.DecisionTreeClassifier(random_state=51,max_depth=None)
clf = clf.fit(X, y1)

#plot da árvore
plt.figure(figsize=(5,5))
tree.plot_tree(clf,feature_names=['Salário','CLT','Tempo'],impurity=False,rounded=True);
print('Importância das features salário, clt e tempo ',clf.feature_importances_)
ye=clf.predict(X)#predição para a amostra de treinamento toda
print('Acurácia na amostra de treinamento ',accuracy_score(y1,ye))
```

```python
#treinamento da árvore
clf1 = tree.DecisionTreeClassifier(random_state=51,max_depth=1)
clf1 = clf1.fit(X, y1)

#plot da árvore
plt.figure(figsize=(5,5))
tree.plot_tree(clf1,feature_names=['Salário','CLT','Tempo'],impurity=False,rounded=True);
print('Importância das features salário, clt e tempo ',clf.feature_importances_)
ye1=clf1.predict(X)#predição para a amostra de treinamento toda
print('Acurácia na amostra de treinamento ',accuracy_score(y1,ye1))
```

```python
#Classificações com os dois modelos

clf.predict([[2500,1,6]]),clf1.predict([[2500,1,6]])

```
