```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
import random as r
from sklearn.metrics import accuracy_score
import math
```

```python
#Cria√ß√£o do dataset
r.seed(1)
X=[]
y=[]

for i in range(1000):
    raio=4.0*r.random()
    ang=6.28*r.random()
    X.append([raio*math.cos(ang),raio*math.sin(ang)])
    y.append(-1)
    
for i in range(1000):
    raio=4.1+4.0*r.random()
    ang=6.28*r.random()
    X.append([raio*math.cos(ang),raio*math.sin(ang)])
    y.append(1)
    
for i in range(len(y)):
    if(y[i]==-1):plt.plot(X[i][0],X[i][1],'bo')
    else:plt.plot(X[i][0],X[i][1],'ro')

plt.show()
```

```python
clf=svm.SVC(kernel='rbf')
clf.fit(X,y)
ye=clf.predict(X)

print('Acur√°cia no treino ',accuracy_score(y,ye))
        

```
