```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
import random as r
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import math
```

```python
#Cria√ß√£o do dataset
r.seed(1)
X=[]
y=[]

for i in range(1000):
    raio=4.0*r.random()
    ang=6.28*r.random()
    X.append([raio*math.cos(ang),raio*math.sin(ang)])
    y.append(-1)
    
for i in range(1000):
    raio=3.0+4.0*r.random()
    ang=6.28*r.random()
    X.append([raio*math.cos(ang),raio*math.sin(ang)])
    y.append(1)
    
for i in range(len(y)):
    if(y[i]==-1):plt.plot(X[i][0],X[i][1],'bo')
    else:plt.plot(X[i][0],X[i][1],'ro')

plt.show()
```

```python
s1=svm.SVC(kernel='rbf')
s1.fit(X,y)
ye=s1.predict(X)

print('Acur√°cia no treino RBF ',accuracy_score(y,ye))
        
```

```python
s2=svm.SVC(kernel='poly', coef0=0.0, gamma=1.0, degree=2)
s2.fit(X,y)
ye2=s2.predict(X)

print('Acur√°cia no treino polinomial ',accuracy_score(y,ye2))
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)

lc=[0.01,0.1,1,10,100]
lg=[0.01,0.1,1,10,100]

c_x=lc[0]
g_x=lg[0]
acur_x=0.5

for c in lc:
    for g in lg:
        clf=svm.SVC(kernel='rbf',C=c,gamma=g)
        clf.fit(X_train,y_train)
        a=accuracy_score(y_test,clf.predict(X_test))
        if a > acur_x:
            acur_x=a
            c_x=c
            g_x=g


print('acur=',acur_x,' c=',c_x,' gamma=',g_x)
    
  
```
